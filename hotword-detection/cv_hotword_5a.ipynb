```python
import pandas as pd

df = pd.read_csv('/content/cv-valid-dev.csv')

hot_words = ["BE CAREFUL", "DESTROY", "STRANGER"]
detected_files = []
```

```python
df.head(5)
```

```python
# Check for hotwords in transcriptions
for index, row in df.iterrows():
    transcription = row['finetuned_text']
    # Convert transcription to string if it's not NaN
    if pd.notna(transcription):

        transcription = str(transcription)
        # print(transcription)
        if any(hot_word in transcription for hot_word in hot_words):
            detected_files.append(row['filename'])
```

```python
print(detected_files)
```

```python
filtered_df = df[df['filename'].isin(detected_files)]
filtered_df
```

```python
# Save the results
with open('detected.txt', 'w') as f:
    for file in detected_files:
        f.write(f"{file}\n")
```
